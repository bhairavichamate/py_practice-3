**find outlier**

```python
# Step-1
def outliers_data(x):

    q1=round(np.quantile(cst_df[x],0.25),2)
    q2=round(np.quantile(cst_df[x],0.50),2)
    q3=round(np.quantile(cst_df[x],0.75),2)

      # Step-2
    IQR=q3-q1

       #step-3
    ub=q3+1.5*IQR

         #Step-4
    lb=q1-1.5*IQR

         #Step-5 and Step-6

    con1=cst_df[x]>ub
    con2=cst_df[x]<lb

          #Step-7 and step-8
    outliers_data=cst_df[con1|con2].values

 

    return outliers_data
```

```python
def outliers(x):
    
    q1=round(np.quantile(cst_df[x],0.25),2)
    q2=round(np.quantile(cst_df[x],0.50),2)
    q3=round(np.quantile(cst_df[x],0.75),2)
    IQR=q3-q1
    ub=q3+1.5*IQR
    lb=q3-1.5*IQR
    con1=cst_df[x]>ub
    con2=cst_df[x]<lb

    outliers=cst_df[con1|con2].values

    return ub,lb
```
